```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./framingham_heart_disease.csv')
```

```python
data.head()
```

```python
data[['totChol','currentSmoker']].head()
```

```python
data[['totChol','currentSmoker']].plot.scatter(x='totChol', y='currentSmoker')
```

```python
w= 0.02
b= -5.6
```

```python
x = np.linspace(0,data['totChol'].max(),100)
y= 1/(1+np.exp(-(w*x+b)))

#grafica de la recta
data.plot.scatter(x='totChol',y='currentSmoker')
plt.plot(x,y,'-r')
plt.ylim(0,data['currentSmoker'].max()*1.1)
#plt.grid()
plt.show()
```
