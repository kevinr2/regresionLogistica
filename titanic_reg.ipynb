```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./DataSet_Titanic.csv')
```

```python
data[['Edad','Sobreviviente']].plot.scatter(x='Edad', y= 'Sobreviviente')
```

```python
w= 0.09
b= -3.6
```

```python
x = np.linspace(0,data['Edad'].max(),100)
y= 1/(1+np.exp(-(w*x+b)))

#grafica de la recta
data.plot.scatter(x='Edad',y='Sobreviviente')
plt.plot(x,y,'-r')
plt.ylim(0,data['Sobreviviente'].max()*1.1)
#plt.grid()
plt.show()
```

```python
data['id'] = 1/(1+np.exp(-(data['Edad']*w+b)))

# calculo de la funcion de error
data['loss_xi'] = -data['Sobreviviente']*np.log(data['id'])-(1-data['Sobreviviente'])*np.log(1-data['id'])
cost_j = data['loss_xi'].mean()
cost_j
```
